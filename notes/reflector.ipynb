```python
import pimms.optics as optics
```

* 建立抛物反射镜，口径1m，焦距10m，开口方向沿z轴（会聚沿-z轴入射的光线），焦点位于坐标系原点。
* 建立单色点光源，单位强度，波长500nm，位于坐标系原点。
* 建立探测器，位于z=10，尺寸1x1，分辨率129x129.

```python
a,p=optics.parabolic_reflector_experiment((0,0,0), (1.,1.,10))
plt.imshow(a,extent=(-.5,.5,-.5,.5))
plt.xlabel('x, in meter')
plt.ylabel('y, in meter')
plt.set_cmap('bone')
cbar=plt.colorbar()
cbar.set_label('amplitude')
```

如上图所示，出射光在z=10处形成直径1m的圆形光斑。

```python
a,p=optics.parabolic_reflector_experiment((0,0,0), (1.,1.,100))
plt.imshow(a,extent=(-.5,.5,-.5,.5))
plt.xlabel('x, in meter')
plt.ylabel('y, in meter')
plt.set_cmap('bone')
cbar=plt.colorbar()
cbar.set_label('amplitude')
```

如上图所示，出射光在z=100处仍然形成直径为1m的圆形光斑，证明出射光为平行光束。
接下来考察出射光波前情况。

```python
a,p=optics.parabolic_reflector_experiment((0,0,0), (1,1,100), resolution=65)
plt.imshow(p,extent=(-.5,.5,-.5,.5))
plt.xlabel('x, in meter')
plt.ylabel('y, in meter')
plt.set_cmap('bwr')
cbar=plt.colorbar()
cbar.set_label('phase')
```

如上图，当光源放置于抛物面焦点时，出射光在探测器上形成等相位光斑。
